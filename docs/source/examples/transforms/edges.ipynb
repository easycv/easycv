# Edges Examples

```python
from easycv import Image
from easycv.io.output import show_grid
from easycv.transforms import Gradient, GradientAngle, GradientMagnitude, Canny
```

For this example we will load an image from the [Dog API](https://dog.ceo/dog-api/). For more examples of image loading check [Image examples](../image.ipynb).

```python
img = Image("https://images.dog.ceo/breeds/sheepdog-english/n02105641_1045.jpg")
img
```

### Gradient

To calculate the gradient of the image we simply apply the Gradient Transform for both axis, for more information check the [reference](../../reference/transforms/edges.rst#easycv.transforms.edges.Gradient). The gradients are displayed in a grid created using [show_grid()](../../reference/io/output.rst#easycv.io.output.show_grid).

```python
grad_x = img.apply(Gradient(axis="x"))
grad_y = img.apply(Gradient(axis="y"))
show_grid([grad_x,grad_y])
```

### Gradient Magnitude

To calculate the magnitude of the image gradient we simply apply the Gradient transform with no axis, for more information check the [reference](../../reference/transforms/edges.rst#easycv.transforms.edges.Gradient).

```python
img.apply(Gradient())
```

### Gradient Angle

To calculate the angles of the image gradient we simply apply the GradientAngle transform, for more information check the [reference](../../reference/transforms/edges.rst#easycv.transforms.edges.GradientAngle).

```python
img.apply(GradientAngle(size=1))
```

### Canny Edge

To get the edges of an image we just apply the Canny transform with the low and high threshold, for more information check the [reference](../../reference/transforms/edges.rst#easycv.transforms.edges.Canny).

```python
img.apply(Canny(low=100, high=200))
```
