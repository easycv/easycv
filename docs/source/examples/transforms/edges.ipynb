# Edges Examples

```python
from easycv import Image
from easycv.io.output import show_grid
from easycv.transforms import Gradient, GradientAngle, Canny, Lines, Circles
```

For this example we will load an image with [random()](../../reference/image.rst#easycv.image.Image.random). For more examples of image loading check [Image examples](../image.ipynb).

```python
img = Image.random()
img
```

### Gradient

To calculate the gradient of the image we simply apply the Gradient Transform for both axis, for more information check the [reference](../../reference/transforms/edges.rst#easycv.transforms.edges.Gradient). The gradients are displayed in a grid created using [show_grid()](../../reference/io/output.rst#easycv.io.output.show_grid).

```python
grad_x = img.apply(Gradient(axis="x"))
grad_y = img.apply(Gradient(axis="y"))
show_grid([grad_x,grad_y])
```

### Gradient Magnitude

To calculate the magnitude of the image gradient we simply apply the Gradient transform without an axis, for more information check the [reference](../../reference/transforms/edges.rst#easycv.transforms.edges.Gradient).

```python
img.apply(Gradient())
```

### Gradient Angle

To calculate the angles of the image gradient we simply apply the GradientAngle transform, for more information check the [reference](../../reference/transforms/edges.rst#easycv.transforms.edges.GradientAngle).

```python
img.apply(GradientAngle(size=1))
```

### Canny Edge

To get the edges of an image we just apply the Canny transform with the low and high threshold, for more information check the [reference](../../reference/transforms/edges.rst#easycv.transforms.edges.Canny).

```python
img.apply(Canny(low=180, high=200))
```

### Lines

To get the lines in an image apply the Lines Transform, for more information check the [reference](../../reference/transforms/edges.rst#easycv.transforms.edges.Lines).

```python
img.apply(Lines(threshold=190))
```

### Circles

To get the circles in an image apply the Circles Transform, for more information check the [reference](../../reference/transforms/edges.rst#easycv.transforms.edges.Circles).

```python
img = Image("https://29comwzoq712ml5vj5gf479x-wpengine.netdna-ssl.com/wp-content/uploads/2019/10/circles-r2.png")
img
```

```python
img.apply(Circles(threshold=70, minDist=10))
```
