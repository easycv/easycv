# Spatial Examples

```python
from easycv import Image
from easycv.transforms import Resize, Rescale, Rotate, Crop, Translate
```

For this example we will load an image with [random()](../../reference/image.rst#easycv.image.Image.random). For more examples of image loading check [Image examples](../image.ipynb).

```python
img = Image.random()
img
```

### Resize

To resize an image we simply apply the Rezise Transform, for more information check the [reference](../../reference/transforms/spatial.rst#easycv.transforms.spatial.Resize).

```python
img.apply(Resize(width=300, height=250))
```

### Rescale

To rescale an image we simply apply the Rescale Transform with the factors of x and y, for more information check the [reference](../../reference/transforms/spatial.rst#easycv.transforms.spatial.Rescale).

```python
img.apply(Rescale(fx=2, fy=1))
```

### Rotate

To rotate an image we simply apply the Rotate Transform with the degrees to rotate, for more information check the [reference](../../reference/transforms/spatial.rst#easycv.transforms.spatial.Rotate).

```python
img.apply(Rotate(degrees=50))
```

### Crop

To crop an image we simply apply the Crop Transform with the boundaries of each side, for more information check the [reference](../../reference/transforms/spatial.rst#easycv.transforms.spatial.Crop).

```python
img.apply(Crop(rectangle=[[50,50],[100,100]]))
```

### Translate

To translate an image we simply apply the Translate Transform with the vector to translate, for more information check the [reference](../../reference/transforms/spatial.rst#easycv.transforms.spatial.Translate).

```python
img.apply(Translate(x=10,y=50))
```
